#### Series
Uma série no pandas é uma estrutura de dados unidimensional que pode conter qualquer tipo de dado.<br>
Semelhante a um vetor, mas com rôtulos para cada elemento.<br>
Como definir uma serie?

```python
#### Series
s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
print(s)
```

#### DataFrame
Um dataframe é uma estrutura de dados rotulados bidimencionais, com linhas e colunas rotuladas, parecidas com uma tabela do excel ou SQL.

```python
import pandas as pd

df = pd.DataFrame({
    'categoria': ['A', 'A', 'B', 'B', 'C'],
    'valor': [10, 20, 30, 40, 50]
})

agrupado = df.groupby('categoria')['valor'].mean()
print(agrupado)
```

```python
# Merge

clientes = pd.DataFrame({
    'id': [1, 2, 3],
    'nome': ['Thiago', 'Judas', 'João']
})
compras = pd.DataFrame({
    'id': [1, 2, 3],
    'valor': [1900, 400, 632]
})

resultado = pd.merge(clientes, compras, on='id', how='inner')
resultado
```

```python
df1 = pd.DataFrame({
    'nome': ['Luana', 'Paula', 'Fabiana']
})
df2 = pd.DataFrame({
    'nome': ['Gabriel', 'João', 'Mikael']
})

resultado = pd.concat([df1, df2])
resultado
```

```python
df = pd.DataFrame({
    'produto': ['A', 'B', 'C', 'A'],
    'loja': ['SP', 'RJ', 'SP', 'RJ'],
    'venda': [100, 200, 150, 250],
    'cidade': ['São Paulo', 'Rio de Janeiro', 'São Paulo',  'Nova Iguaçu']
})
tabela = df.pivot_table(values='venda', index='produto', columns=['loja', 'cidade'], aggfunc='sum')
tabela
```

```python
df.memory_usage(deep=True)
```

```python
# Realizando a manipulação dos dados e dobrando o valor das idades.
df['idade_dobrada'] = df['idade'].apply(lambda x: x * 2)
```

### Etapa 3: Limpeza e Preparação dos Dados
No pandas existem diversas funções para lidar com dados fora de ordem, sem que seja necessário realizar tratamentos específicos..<br>
Porém, outras manipulações manuais podem ser feitas para melhor atender as necessidades

```python
df = pd.DataFrame({'A': [1, 2, np.nan, 4],
                   'B': [5, np.nan, 7, 8],
                   'C': [np.nan, 10, 11, 12]})
```

```python
# isnull()
# Retorna false caso um valor não seja nulo, ou true se for nulo.
df.isnull()
```

```python
# Pode ser usado para verificar a quantidade de nulos em uma coluna específica, mas se torna inútil, visto que info já faz isso.
df['A'].isnull().sum()
```

```python
# isnull pode ser usado para identificar os indices dos valores nulos.
df.loc[df['A'].isnull()].head()
```

```python
# Preenchendo nulos com 0
df.fillna(0)
```
